```
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random


nodes = range(50)

edges = [(random.choice(nodes), random.choice(nodes)) for edge_id in range(170)]

G = nx.Graph()
G.add_edges_from(edges)

zeros = [0.0 for i in range(40)]
ones = [1.0 for i in range(10)]
labels = zeros + ones
random.shuffle(labels)
print(labels)

val_map = {nodes[i]:labels[i] for i in range(len(nodes))}
print(val_map)

values = [val_map.get(node) for node in G.nodes()]

nx.draw(G, cmap=plt.get_cmap('coolwarm'), node_color=values ,font_color='white', node_size=70)
plt.show()
```
